## Ejercicio 1

Contar cuántos valores hay en una lista de enteros. Contar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
contador = 0
#Respuesta:
#Debe dar 5
for numero in nums:
  contador += 1
  if numero == 0:
     break

print(contador)
```

## Ejercicio 2

Sumar los valores que hay en una lista de enteros. Sumar solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
#Respuesta:
#Debe dar 21
for valores in nums:
  suma = suma +valores
  if valores == 0:
    break
print(suma)
```

## Ejercicio 3

Calcular el valor medio de los valores que hay en una lista de enteros. Calcular solo hasta el número 0.

```python
nums = [6,8,5,2,0,6,5]
suma = 0
contador = 0
#Respuesta:
#Debe dar 4.2
for valores in nums:
  contador = contador+1
  suma = suma +valores
  if valores == 0:
    break
media = suma / contador

print(media)
```

## Ejercicio 5 (Opcional)

Crear una lista a partir de una secuencia de valores de entrada terminada en *.
Una vez creada, modificarla elevando al cuadrado todos los pares y al cubo todos los impares, y mostrarla.

```python
lista = [3,56,6,7,"*",2,78]
lista_2 = []
#Respuesta:
#Debe dar [27, 3136, 36, 343]
for valores in lista:
  try:
      numero = int(valores)
      if numero % 2 == 0 :
        lista_2.append (numero**2)
      else :
        lista_2.append(numero**3)
  except ValueError:
    break
print(lista_2)


print(lista)
```
